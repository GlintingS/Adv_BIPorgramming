```python
def num_gen(start, stop):
    if start > stop:
        start, stop = stop, start

    for _ in range(10):
        yield random.randint(start, stop)
```

```python
import functools
import logging

# Configure logging
logging.basicConfig(level=logging.ERROR, format='%(asctime)s - %(levelname)s - %(message)s')

def catch_errors(func):
    """Decorator to catch specific errors and log them."""
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except (ValueError, TypeError, IndexError) as e:
            # Log the exception details
            logging.error(f"Error in {func.__name__}: {e}", exc_info=False)
            return None # Or return a default value like []
    return wrapper

# --- Usage Example ---

@catch_errors
def process_data(data):
    return data[10] # Potential IndexError

@catch_errors
def convert_to_int(value):
    return int(value) # Potential ValueError/TypeError

# The program will continue to run
print("Program starting...")
process_data([1, 2])       # Logs an error, returns None
convert_to_int("not_an_int") # Logs an error, returns None
print("Program finished successfully.")
```
