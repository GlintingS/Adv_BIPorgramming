```python
import random
import functools
import operator
import logging
from functools import reduce
```

```python
# Configure logging
logging.basicConfig(level=logging.ERROR, format='%(asctime)s - %(levelname)s - %(message)s')

def catch_errors(func):
    """Decorator to catch specific errors and log them."""
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        while True:
            try:
                return func(*args, **kwargs)
            except (ValueError, TypeError, IndexError) as e:
                # Log the exception details
                logging.error(f"Error in {func.__name__}: {e}", exc_info=False)
                print(f"Invalid input: {e}. Please try again.")
            except Exception as e:
                logging.error(f"Unexpected error in {func.__name__}: {type(e).__name__} - {e}")
                print(f"Invalid input: {e}. Please try again.")
    return wrapper
```

```python
@catch_errors
def num_gen(start, stop):
    if not (isinstance(start, int) and isinstance(stop, int)):
        raise TypeError("Start and stop must be integers.")
    if start >= stop:
        raise IndexError("Start must be less than stop.")
    if stop < 0 or start < 0:
        raise ValueError("Start and stop must be non-negative.")

    for _ in range(10):
        yield random.randint(start, stop)
```

```python
def num_trans(start, stop):
    num = num_gen(start, stop)
    my_num = list(num)

    mapped_num = map(lambda x: x**2, my_num)
    mapped_list = list(mapped_num)

    filtered_list = list(filter(lambda x: x > 10, mapped_list))

    sum_result = reduce(operator.add, filtered_list)

    return my_num, mapped_list, filtered_list, sum_result
```

```python
if __name__ == "__main__":
    try:
        start = 5
        stop = 19
        my_num, mapped_list, filtered_list, sum_result = num_trans(start, stop)

        print(my_num)
        print(mapped_list)
        print(filtered_list)
        print(sum_result)
    except Exception as e:
        logging.error(f"pipeline error: {type(e).__name__} - {e}")

    try:
        start = -5
        stop = 19
        my_num, mapped_list, filtered_list, sum_result = num_trans(start, stop)

        print(my_num)
        print(mapped_list)
        print(filtered_list)
        print(sum_result)
    except Exception as e:
        logging.error(f"Error: {type(e).__name__} - {e}")

    try:
        start = 35
        stop = 19
        my_num, mapped_list, filtered_list, sum_result = num_trans(start, stop)

        print(my_num)
        print(mapped_list)
        print(filtered_list)
        print(sum_result)
    except Exception as e:
        logging.error(f"Error: {type(e).__name__} - {e}")

    try:
        start = 5.78
        stop = 19
        my_num, mapped_list, filtered_list, sum_result = num_trans(start, stop)

        print(my_num)
        print(mapped_list)
        print(filtered_list)
        print(sum_result)
    except Exception as e:
        logging.error(f"Error: {type(e).__name__} - {e}")
```
